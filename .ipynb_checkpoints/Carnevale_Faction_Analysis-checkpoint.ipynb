```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.set() # Setting seaborn as default style even if use only matplotlib
%matplotlib inline
pd.set_option('display.max_columns', None)
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
```

```python
df = pd.read_csv('factions.csv')
```

```python
#df.groupby('faction').describe()
```

**Mittelwert, Standardabweichung, Minimal- und Maximalwerte der stats je nach Fraktion**

```python
analysiscolumns='actions','life','will','command','ducats','movement','dexterity','attack','protection','mind'
analysisvalues='mean', 'std', 'min', 'max'
df.groupby('faction')[analysiscolumns].describe().applymap('{:,.2f}'.format).loc[:,(slice(None),['mean', 'std', 'min', 'max'])].transpose()
```

**Analyse der stats weiter unterteilt nach Typ (Hero, Leader, Henchman)**

```python
df.groupby(['faction','type'])[analysiscolumns].describe().applymap('{:,.2f}'.format).loc[:,(slice(None),['mean', 'std', 'min', 'max'])].transpose()
```

```python
for i in analysiscolumns:
    sns.boxplot(data=df, x=df['faction'], y=df[i], hue = df['type'], palette = 'bone')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
    plt.show()
```

```python
for i in analysiscolumns:
    sns.boxplot(data=df, x=df['faction'], y=df[i], palette = 'bone')
    plt.show()
```

```python
df['abilities'] = df['character abilities'].str.lower().str.split(',')
#df['abilities'].str.strip()

def strip_element(my_list):
    return [x.strip() for x in my_list]

df['abilities'] = df['abilities'].apply(strip_element)
```

```python
valuecounts=df.explode('abilities').groupby('faction')['abilities'].value_counts()
print(valuecounts.to_string())
```

```python
valuecountstransposed=df.explode('abilities').groupby('abilities')['faction'].value_counts()
print(valuecountstransposed.to_string())
```
